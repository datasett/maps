```python
import pandas as pd
```

```python
df = pd.read_excel('https://www.bring.no/radgivning/sende-noe/adressetjenester/postnummer/_/attachment/download/394005e4-9559-4403-ae45-5c8216ecc4c9:ee83f7ab4f5ed6d3aed3c15c3491b561cb4d520f/Nye,%20endrede%20og%20opph%C3%B8rte%20postnummer.xlsx',
                  converters={'Kommunenr':str,'Kommunenr_etter':str, 'Postnummer_etter':str})
```

```python
df.head()
```

```python
df_kommune = df[['Kommunenr', 'Kommune']].drop_duplicates()
df_kommune_etter = df[['Kommunenr_etter', 'Kommune_etter']].drop_duplicates()
df_kommune_etter.columns=['Kommunenr', 'Kommune']
df_kommune = df_kommune.append(df_kommune_etter, ignore_index=True).drop_duplicates()
df_kommune['Kommunenr'] = df_kommune['Kommunenr'].str.zfill(4)
df_kommune['Fylkenr'] = df_kommune['Kommunenr'].str[0:2]
```

```python
df_kommune.head()
```

```python
df_kommune['Fylkenr'].unique()
```
